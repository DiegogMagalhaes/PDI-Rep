## 1. Geração de um checkboard usando a biblioteca Numpy

### A.

```python
B = np.zeros((10,10))
W = np.ones((10,10))

print(B)
print(W)
```

### B.

```python
BW = np.concatenate((B, W), axis=1, out=None, dtype=None, casting="same_kind")
plt.imshow(BW, cmap='gray')
```

### C. 

```python
BW_tiled = np.tile(BW, 5)
plt.imshow(BW_tiled, cmap='gray')
```

### D.

```python
WB = np.concatenate((W, B), axis=1)
plt.imshow(WB, cmap='gray')
```

```python
WB_tiled = np.tile(WB, 5)
plt.imshow(WB_tiled, cmap='gray')
```

### E.

```python
checkboard = np.concatenate((BW_tiled, WB_tiled), axis=0, out=None, dtype=None, casting="same_kind")
checkboard_100 = np.tile(checkboard,(5,1))
plt.imshow(checkboard_100, cmap='gray')
```

## 2. Geração de um checkboard cinza apartir do checkboard preto e branco

### Percorremos pela imagem buscando pelos seus valores brancos, alterando seu valor para cinza

```python
checkboard_100_gray = checkboard_100.astype('uint8')

i=0
j=0

while(i<100):
    while(j<100):

        if(checkboard_100_gray[i][j]==1):
            checkboard_100_gray[i][j]=127

        j+=1

    i+=1
    j=0

plt.imshow(checkboard_100_gray,cmap='gray',vmin =0, vmax=255)
plt.title("imagem cinza")
```

## 3. Geração de novos checkboards nas cores: vermelho, verde, azul, amarelo e magenta.

### Alteração de diminsão do checkboard para 3 dimensões

```python
checkboard_100_gray = np.array(checkboard_100_gray, dtype=np.uint8, order='F')
checkboard_100_gray.resize(100, 100, 3)
```

### Declaração dos novos tabuleiros, cores, mascaras de cores

```python
white = [255, 255, 255]
red = [255, 0, 0]
green = [0, 255, 0]
blue = [0, 0, 255]
yellow = [255, 255, 0]
magenta = [255, 0, 255]

checkboard_100_red = np.array(checkboard_100_gray, dtype=np.uint8)
checkboard_100_green = np.array(checkboard_100_gray, dtype=np.uint8)
checkboard_100_blue = np.array(checkboard_100_gray, dtype=np.uint8)
checkboard_100_yellow = np.array(checkboard_100_gray, dtype=np.uint8)
checkboard_100_magenta = np.array(checkboard_100_gray, dtype=np.uint8)

green_mask = red_mask = blue_mask = yellow_mask = magenta_mask = checkboard_100_gray[:, :, 0] == 127
```

### Os novos checkboards receberão como indice as mascaras declaradas previamente e receberão as cores RGBs declaradas

```python
checkboard_100_red[red_mask] = red
checkboard_100_green[green_mask] = green
checkboard_100_blue[blue_mask] = blue
checkboard_100_yellow[yellow_mask] = yellow
checkboard_100_magenta[magenta_mask] = magenta
```

## Exibição

```python
plt.subplot(2, 3, 1)
plt.title("White")
plt.imshow(checkboard_100, cmap='gray')

plt.subplot(2, 3, 2)
plt.title("Red")
plt.imshow(checkboard_100_red, vmin = 0, vmax = 255)

plt.subplot(2, 3, 3)
plt.title("Green")
plt.imshow(checkboard_100_green, vmin = 0, vmax = 255)

plt.subplot(2, 3, 4)
plt.title("Blue")
plt.imshow(checkboard_100_blue, vmin = 0, vmax = 255)

plt.subplot(2, 3, 5)
plt.title("Yellow")
plt.imshow(checkboard_100_yellow, vmin = 0, vmax = 255)

plt.subplot(2, 3, 6)
plt.title("Magenta")
plt.imshow(checkboard_100_magenta, vmin = 0, vmax = 255)

# Exibição
plt.show()
```
