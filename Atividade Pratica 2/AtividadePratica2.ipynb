### Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Equalizando a primeira imagem

```python
img_0 = cv2.imread('lena_gray_512.tif',0)
img_0_eq = cv2.equalizeHist(img_0)

fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.add_subplot(1,2,1)
plt.title("Imagem")
plt.imshow(img_0, cmap='gray')
fig.add_subplot(1,2,2)
plt.title("Imagem Equalizada")
plt.imshow(img_0_eq, cmap='gray')
 
```

### Histograma da primeira imagem

```python
hist_0 = cv2.calcHist([img_0],[0],None,[256],[0,256]) 
hist_0_eq = cv2.calcHist([img_0_eq],[0],None,[256],[0,256]) 


fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(8)

plt.subplot(2, 2, 1)
plt.title("Histograma")
plt.plot(hist_0)

plt.subplot(2, 2, 2)
plt.title("Imagem")
plt.imshow(img_0,cmap='gray')

plt.subplot(2, 2, 3)
plt.title("histograma Equalizado")
plt.plot(hist_0_eq)

plt.subplot(2, 2, 4)
plt.title("Imagem Equalizada")
plt.imshow(img_0_eq,cmap='gray')

```

## Equalizando a quarta imagem

```python
img_3 = cv2.imread('20 de mar. de 2024 as 16_12 (1).tiff',1)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB) 

ycrcb_img_3 = cv2.cvtColor(img_3, cv2.COLOR_RGB2YCrCb)
ycrcb_img_3[:, :, 0] = cv2.equalizeHist(ycrcb_img_3[:, :, 0])

hsv_img_3 = cv2.cvtColor(img_3, cv2.COLOR_RGB2HSV)
hsv_img_3[:, :, 0] = cv2.equalizeHist(hsv_img_3[:, :, 0])


img_3_eq_y = cv2.cvtColor(ycrcb_img_3, cv2.COLOR_YCrCb2RGB)
img_3_eq_v = cv2.cvtColor(hsv_img_3, cv2.COLOR_HSV2RGB)

fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(10)
fig.add_subplot(1,3,1)
plt.title("Imagem")
plt.imshow(img_3, vmin=0, vmax=255)
fig.add_subplot(1,3,2)
plt.title("Imagem Equalizada em Y")
plt.imshow(img_3_eq_y, vmin=0, vmax=255)
fig.add_subplot(1,3,3)
plt.title("Imagem Equalizada em V")
plt.imshow(img_3_eq_v, vmin=0, vmax=255)
```

### Histograma da quarta imagem

```python
hist_3_r = cv2.calcHist([img_3],[0],None,[256],[0,256])
hist_3_g = cv2.calcHist([img_3],[1],None,[256],[0,256])
hist_3_b = cv2.calcHist([img_3],[2],None,[256],[0,256])

hist_3_eq_r = cv2.calcHist([img_3_eq_y],[0],None,[256],[0,256]) 
hist_3_eq_g = cv2.calcHist([img_3_eq_y],[1],None,[256],[0,256]) 
hist_3_eq_b = cv2.calcHist([img_3_eq_y],[2],None,[256],[0,256]) 


fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(20)

plt.subplot(2, 5, 1)
plt.title("Histograma em R")
plt.plot(hist_3_r, color='r')

plt.subplot(2, 5, 2)
plt.title("Histograma em G")
plt.plot(hist_3_g, color='g')

plt.subplot(2, 5, 3)
plt.title("Histograma em B")
plt.plot(hist_3_b, color='b')

plt.subplot(2, 5, 4)
plt.title("Histograma em RGB")
plt.plot(hist_3_r, color='r')
plt.plot(hist_3_g, color='g')
plt.plot(hist_3_b, color='b')

plt.subplot(2, 5, 5)
plt.title("Imagem")
plt.imshow(img_3, vmin=0, vmax=255)

plt.subplot(2, 5, 6)
plt.title("Histograma em R")
plt.plot(hist_3_eq_r, color='r')

plt.subplot(2, 5, 7)
plt.title("Histograma em G")
plt.plot(hist_3_eq_g, color='g')

plt.subplot(2, 5, 8)
plt.title("Histograma em B")
plt.plot(hist_3_b, color='b')

plt.subplot(2, 5, 9)
plt.title("Histograma em RGB")
plt.plot(hist_3_eq_r, color='r')
plt.plot(hist_3_eq_g, color='g')
plt.plot(hist_3_eq_b, color='b')

plt.subplot(2, 5, 10)
plt.title("Imagem")
plt.imshow(img_3_eq_y, vmin=0, vmax=255)
```
