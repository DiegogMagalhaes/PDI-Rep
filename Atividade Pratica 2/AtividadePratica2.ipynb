### Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Algoritmo

```python
def equalizeHist2(imagem,x,y):
   
    hist_0 = cv2.calcHist([imagem],[0],None,[256],[0,256]) 
    i = 0    
    
    while(i<255):
        hist_0[i] = hist_0[i]/(x*y)  
        i+=1 

    probabilidade_acumulada = 0    

    hist_02 = hist_0.copy()

    k = 255

    j = 0

    while(j<k):       

        probabilidade_acumulada = probabilidade_acumulada + hist_0[j]        
        hist_02[j] = probabilidade_acumulada
        j+=1

    #mudo a intensidade de cada pixel da imagem
    #agora com a probabilidade acumulada de cada intensidade usamos a fÃ³rmula do slide para modificar a intensidade dos pixels da imagem

    i = 0
    i2 = 0

    #percorrendo no eixo x e no eixo y
    while(i<x):
        while(i2<y):
            imagem[i][i2] = 255 * hist_02[imagem[i][i2]][0]            
            i2+=1
        i+=1
        i2=0

    return imagem
 
```

## Equalizando a primeira imagem

```python
img_0 = cv2.imread('lena_gray_512.tif',0)
img_02 = cv2.imread('lena_gray_512.tif',0)

img_0_eq2 = equalizeHist2(img_02,512,512)        

fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.add_subplot(1,2,1)
plt.title("Imagem")
plt.imshow(img_0, cmap='gray')
fig.add_subplot(1,2,2)
plt.title("Imagem Equalizada")
plt.imshow(img_0_eq2, cmap='gray')
 
```

### Histograma da primeira imagem

```python
hist_0 = cv2.calcHist([img_0],[0],None,[256],[0,256]) 
hist_0_eq = cv2.calcHist([img_0_eq2],[0],None,[256],[0,256]) 


fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(8)

plt.subplot(2, 2, 1)
plt.title("Histograma")
plt.plot(hist_0)

plt.subplot(2, 2, 2)
plt.title("Imagem")
plt.imshow(img_0,cmap='gray')

plt.subplot(2, 2, 3)
plt.title("histograma Equalizado")
plt.plot(hist_0_eq)

plt.subplot(2, 2, 4)
plt.title("Imagem Equalizada")
plt.imshow(img_0_eq2,cmap='gray')

```
